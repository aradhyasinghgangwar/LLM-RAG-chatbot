```python
from langchain import FAISS

#operations on text file
import langchain.embeddings
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.prompts import ChatPromptTemplate
from langchain.schema import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain.vectorstores import FAISS
from langchain_community.llms import HuggingFaceHub
```

```python
def format_docs(docs):
      return "\n\n".join([d.page_content for d in docs])
```

```python
def chat_with_rag(message):
  loader = TextLoader(r"/content/output.txt")
  full_text= loader.load()
  str_list = [str(num) for num in full_text]
  single_string = " ".join(str_list)
  #chunking

  text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
  texts = text_splitter.split_text(single_string)
  #embedding the data
  embeddings = HuggingFaceEmbeddings()
  #db = Chroma.from_texts(texts, embeddings)
  db = FAISS.from_texts(texts, embeddings)


  #Retriever from the database
  retriever = db.as_retriever()
  #Zephyr-7b-beta
  llm = HuggingFaceHub(
    repo_id="HuggingFaceH4/zephyr-7b-beta",
    task="text-generation",
    model_kwargs={
        "max_new_tokens": 512,
        "top_k": 30,
        "temperature": 0.1,
        "repetition_penalty": 1.03,
    },
    huggingfacehub_api_token= hg_key,# Replace with your actual huggingface token
)

  template = """Answer the question based only on the following context:

    {context}

    Question: {question}
    """
  prompt = ChatPromptTemplate.from_template(template)
  model = llm
  chain = (
        {"context": retriever | format_docs, "question": RunnablePassthrough()}
        | prompt
        | model
        | StrOutputParser()
    )
  return chain.invoke(message)

```
